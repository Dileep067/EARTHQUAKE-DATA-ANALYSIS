```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset

df = pd.read_csv("/content/database.csv")
df.head()
```

```python
# Check basic info and missing values
df.info()
df.isnull().sum()
```

```python
# Show summary statistics
df.describe()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Type', data=df)
plt.title("Count of Earthquake Types")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Magnitude'], bins=20, color='skyblue')
plt.title("Earthquake Magnitude Distribution")
plt.xlabel("Magnitude")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6, 6))
df['Status'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#99ff99'])
plt.title("Earthquake Data Status")
plt.ylabel('')
plt.show()
```

```python
top_quakes = df.sort_values(by='Magnitude', ascending=False).head(10)
top_quakes[['Date', 'Time', 'Magnitude', 'Latitude', 'Longitude']]
```

```python
# Convert 'Date' to datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Extract year from date
df['Year'] = df['Date'].dt.year

# Plot count of earthquakes per year
plt.figure(figsize=(12, 5))
sns.countplot(x='Year', data=df, palette='viridis')
plt.xticks(rotation=90)
plt.title("Number of Earthquakes by Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Depth'], bins=30, color='coral')
plt.title("Distribution of Earthquake Depths")
plt.xlabel("Depth (km)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
# Drop unnecessary columns with many missing values or low importance
df_cleaned = df.drop(columns=[
    'Depth Error', 'Depth Seismic Stations', 'Magnitude Error',
    'Magnitude Seismic Stations', 'Azimuthal Gap', 'Horizontal Distance',
    'Horizontal Error', 'Root Mean Square', 'Location Source', 'Magnitude Source'
])

df_cleaned.head()
```

```python
# Convert 'Date' and 'Time' to datetime
df_cleaned['Datetime'] = pd.to_datetime(df_cleaned['Date'] + ' ' + df_cleaned['Time'], errors='coerce')

# Drop rows where datetime couldn't be parsed
df_cleaned = df_cleaned.dropna(subset=['Datetime'])

# Set datetime as index
df_cleaned.set_index('Datetime', inplace=True)

df_cleaned[['Latitude', 'Longitude', 'Depth', 'Magnitude']].describe()
```

```python
# Round magnitude for grouping
df['Rounded Magnitude'] = df['Magnitude'].round()

plt.figure(figsize=(8, 5))
sns.countplot(x='Rounded Magnitude', data=df, palette='Set2')
plt.title("Earthquakes by Rounded Magnitude")
plt.xlabel("Magnitude")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df_cleaned['Magnitude'], bins=30, kde=True, color='orange')
plt.title("Distribution of Earthquake Magnitudes")
plt.xlabel("Magnitude")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df_cleaned, x='Longitude', y='Latitude',
                hue='Magnitude', palette='coolwarm', size='Magnitude', legend=False)
plt.title("Earthquake Locations by Magnitude")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.show()
```

```python
monthly_quakes = df_cleaned.resample('M').size()

plt.figure(figsize=(14, 5))
monthly_quakes.plot()
plt.title("Number of Earthquakes Over Time (Monthly)")
plt.xlabel("Date")
plt.ylabel("Number of Earthquakes")
plt.grid(True)
plt.show()
```

```python
monthly_quakes = df_cleaned.resample('M').size()

plt.figure(figsize=(14, 5))
monthly_quakes.plot()
plt.title("Number of Earthquakes Over Time (Monthly)")
plt.xlabel("Date")
plt.ylabel("Number of Earthquakes")
plt.grid(True)
plt.show()

```

```python
# Top 10 most frequent location sources
top_locations = df['Location Source'].value_counts().head(10)

plt.figure(figsize=(8, 5))
top_locations.plot(kind='bar', color='teal')
plt.title("Top 10 Earthquake Reporting Locations")
plt.xlabel("Location Source")
plt.ylabel("Number of Reports")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Just take first 100 rows to keep it readable
sample_df = df[['Depth', 'Magnitude']].head(100)

plt.figure(figsize=(10, 4))
plt.plot(sample_df['Depth'], label='Depth (km)', color='blue')
plt.plot(sample_df['Magnitude'], label='Magnitude', color='red')
plt.title("Depth and Magnitude Comparison (First 100 Earthquakes)")
plt.xlabel("Event Index")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```
